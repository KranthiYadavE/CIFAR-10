```python
import numpy as np
import pandas as pd
import torch
from transformers import DistilBertTokenizerFast, DistilBertForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset # load_metric has been removed from here
from evaluate import load # this will give you access to load_metric
from sklearn.model_selection import train_test_split
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
import emoji
import re
```

```python
# Load the dataset directly from Hugging Face
dataset = load_dataset("go_emotions")

# Convert to pandas DataFrame for easier cleaning
# 2. Use the existing splits directly
train_dataset = dataset['train']
val_dataset = dataset['validation']
test_dataset = dataset['test']
```

```python
def cleaned_text(text):
    # Convert emojis to text
    text = emoji.demojize(text)

    # Remove URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Remove special characters except basic punctuation
    text = re.sub(r'[^\w\s.,!?]', '', text)

    # Tokenize and join to normalize spacing
    tokens = word_tokenize(text)
    return ' '.join(tokens)

# Apply cleaning
train_dataset = train_dataset.map(lambda x: {'cleaned_text': cleaned_text(x['text'])})
val_dataset = val_dataset.map(lambda x: {'cleaned_text': cleaned_text(x['text'])})
test_dataset = test_dataset.map(lambda x: {'cleaned_text': cleaned_text(x['text'])})
```

```python
# Tokenize using the original splits
tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')

def tokenize_function(batch):
    return tokenizer(batch['cleaned_text'], padding='max_length', truncation=True)

tokenized_train = train_dataset.map(tokenize_function, batched=True)
tokenized_val = val_dataset.map(tokenize_function, batched=True)
tokenized_test = test_dataset.map(tokenize_function, batched=True)

# Prepare labels (assuming we're using the first label only)
def prepare_labels(batch):
    return {'labels': batch['labels'][0] if batch['labels'] else 27}  # 27=neutral

tokenized_train = tokenized_train.map(prepare_labels)
tokenized_val = val_dataset.map(prepare_labels)
tokenized_test = test_dataset.map(prepare_labels)
```

```python
import numpy as np

def convert_to_multihot(example):
    # Initialize 28-dimensional zero vector
    labels = np.zeros(28, dtype=np.float32)

    # Handle both cases: single label (27) or list of labels ([1,5,27])
    label_indices = example['labels']
    if isinstance(label_indices, int):  # If single label
        label_indices = [label_indices]

    # Set 1s for all present labels
    for idx in label_indices:
        if idx < 28:  # Ensure valid index
            labels[idx] = 1.0

    return {'labels': labels}

# Apply conversion
tokenized_train = tokenized_train.map(convert_to_multihot)
tokenized_val = tokenized_val.map(convert_to_multihot)
tokenized_test = tokenized_test.map(convert_to_multihot)

# Verify
print("Converted labels example:", tokenized_train[0]['labels'])
# Should output something like: [0. 0. ... 1. 0.] (length 28)
```

```python
from transformers import DistilBertForSequenceClassification, Trainer, TrainingArguments
import torch

# Multi-label configuration
model = DistilBertForSequenceClassification.from_pretrained(
    "distilbert-base-uncased",
    num_labels=28,
    problem_type="multi_label_classification"
)

# GPU setup
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    # Apply sigmoid and threshold for multi-label classification
    preds = torch.sigmoid(torch.tensor(logits)) > 0.5
    return {
        'micro_f1': f1_score(labels, preds, average='micro'),
        'macro_f1': f1_score(labels, preds, average='macro'),
        'accuracy': accuracy_score(labels, preds)
    }

training_args = TrainingArguments(
    output_dir="./emotivechat_results",
    eval_strategy="epoch",
    learning_rate=3e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=4,
    weight_decay=0.01,
    save_strategy="epoch",
    load_best_model_at_end=True,
    metric_for_best_model="micro_f1",
    logging_steps=50
)
```

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_train,
    eval_dataset=tokenized_val,
    compute_metrics=compute_metrics
)
```

```python
print("Starting training...")
train_results = trainer.train()
print(f"Training completed! Final metrics: {train_results.metrics}")
```
